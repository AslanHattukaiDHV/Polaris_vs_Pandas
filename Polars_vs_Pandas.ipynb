### Measures

We are interested in 3 diffrent measures:

1. `Runtime` to measure runtime we will be using the time library `time.time()` 

2. `CPU usage`to measure the percentage of CPU used we will use the  psutil library `psutil.cpu_percent()`

3. `Memory usage` to measure the memory used we will need to adjust an existing comand from the psutil library `measure_memory()`*<font style="color:RED">see cell below</font>*

By running these commands before and after the executing of a function and substracting the pre execution value from the post execution value we will end up with the `time`, `%cpu`and `mb` used during the execution of the function. 

Doing this for the same operations written using Pandas and Polars will allow us to compare the values between the two libraries. To reduce the amount of outlier measures we will be taking the avaerage of these values from a number of runs.

```python
def measure_memory():

    process = psutil.Process() # Creates objec that represents current memory usage
    memory_usage = process.memory_info().rss  # Gets the RSS (Resident set size) in bytes
    memory_usage_mb = memory_usage / (1024 * 1024)  # Convert bytes to megabytes
    
    return memory_usage_mb
```

```python
#the aggreagate functions we will use are
aggregate_functions = ['min', 'max', 'mean', 'median', 'std']
```

```python
# adjust so data is saved to a file

# Select columns to include in the aggregate functions
df = pd.read_csv("pd_agr_dfc27r1000000.csv")
included_columns = df.columns[7:]

# Initialize lists to store the runtimes and CPU usages
runtimes_agr_pd_C = []
cpu_usages_agr_pd_C = []
memory_usages_agr_pd_C = []

# Perform the aggregate functions and record the runtime and CPU usage
for function in aggregate_functions:
    function_runtimes = []
    function_cpu_usages = []
    function_memory_usages = []
    
    for _ in range(1000):
        start_time = time.time()
        cpu_usage_start = psutil.cpu_percent()
        memory_usage_start = psutil.Process().memory_percent()
        
        result = df[included_columns].agg(function)
    
        end_time = time.time()
        cpu_usage_end = psutil.cpu_percent()
        memory_usage_end = psutil.Process().memory_percent()
    
        runtime = end_time - start_time
        cpu_usage = cpu_usage_end - cpu_usage_start
        memory_usage = memory_usage_end - memory_usage_start
    
        function_runtimes.append(runtime)
        function_cpu_usages.append(cpu_usage)
        function_memory_usages.append(memory_usage)
    
    average_runtime = sum(function_runtimes) / len(function_runtimes)
    average_cpu_usage = sum(function_cpu_usages) / len(function_cpu_usages)
    average_memory_usage = sum(function_memory_usages) / len(function_memory_usages)
    
    runtimes_agr_pd_C.append(average_runtime)
    cpu_usages_agr_pd_C.append(average_cpu_usage)
    memory_usages_agr_pd_C.append(average_memory_usage)
    
    print(f"Aggregate function: {function}")
    print(result)
    print(f"Average Runtime: {average_runtime:.6f} seconds")
    print(f"Average CPU Usage: {average_cpu_usage:.3f}%")
    print(f"Average memory Usage: {average_memory_usage:.3f}%\n")
    
runtimes_agr_pd_C = [round(usage, 6) for usage in runtimes_agr_pd_C]
cpu_usages_agr_pd_C = [round(usage, 3) for usage in cpu_usages_agr_pd_C]
memory_usages_agr_pd_C = [round(usage, 3) for usage in memory_usages_agr_pd_C]

# Save the runtimes and CPU usages to variables
min_runtime, max_runtime, mean_runtime, median_runtime, std_runtime = runtimes_agr_pd_C
min_cpu_usage, max_cpu_usage, mean_cpu_usage, median_cpu_usage, std_cpu_usage = cpu_usages_agr_pd_C
min_memory_usage, max_memory_usage, mean_memory_usage, median_memory_usage, std_memory_usage = memory_usages_agr_pd_C
```
