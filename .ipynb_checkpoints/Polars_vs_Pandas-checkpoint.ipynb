```python
import time # measure runtime
import psutil # measure cpu usage
import string 
import random
import numpy as np
import pandas as pd
print(f"pandas version used here is {pd.__version__}.")
import polars as pl
print(f"polars version used here is {pl.__version__}.")
from IPython.display import Image
import seaborn as sns
import matplotlib.pyplot as plt
```

### <font style="color:red">Test data</font> 

<font style="color:black">The first step to comparing the two libraries will be generating a suitable table to use. We will be working with a CRM type of table. In order to see how the libraries handle larger amounts of data we will be scaling the table horizontally by adding columns and vertically by adding more rows.</font> 

```python
# This will be a eurpean CRM
european_countries = [
    'Albania', 'Andorra', 'Armenia', 'Austria', 'Azerbaijan', 'Belarus', 'Belgium', 'Bosnia and Herzegovina',
    'Bulgaria', 'Croatia', 'Cyprus', 'Czech Republic', 'Denmark', 'Estonia', 'Finland', 'France', 'Georgia',
    'Germany', 'Greece', 'Hungary', 'Iceland', 'Ireland', 'Italy', 'Kazakhstan', 'Kosovo', 'Latvia', 'Liechtenstein',
    'Lithuania', 'Luxembourg', 'Malta', 'Moldova', 'Monaco', 'Montenegro', 'Netherlands', 'North Macedonia',
    'Norway', 'Poland', 'Portugal', 'Romania', 'Russia', 'San Marino', 'Serbia', 'Slovakia', 'Slovenia', 'Spain',
    'Sweden', 'Switzerland', 'Turkey', 'Ukraine', 'United Kingdom', 'Vatican City'
]
```

```python
# This function was used to generate several dfs these will be used to test the libraries against increaslinly larger tables

num_rows = 100

# Generate CRM data
data = {
    'CustomerID': range(1, num_rows + 1),
    'Name': [''.join(random.choices(string.ascii_letters, k=5)) for _ in range(num_rows)],
    'Email': [''.join(random.choices(string.ascii_lowercase, k=7)) + '@example.com' for _ in range(num_rows)],
    'Phone': [''.join(random.choices(string.digits, k=10)) for _ in range(num_rows)],
    'Address': [''.join(random.choices(string.ascii_letters + string.digits, k=10)) for _ in range(num_rows)],
    'Country': [random.choice(european_countries) for _ in range(num_rows)],
    'Gender': [random.choice(['Male', 'Female']) for _ in range(num_rows)],
}

# Add random integer columns
for i in range(1, 6): # adjust range to increase amount of cols
    column_name = f'Int{i}'
    data[column_name] = random.choices(range(100), k=num_rows)

# Add random float columns
for i in range(1, 6): # adjust range to increase amount of cols
    column_name = f'Float{i}'
    data[column_name] = [random.uniform(0, 1) for _ in range(num_rows)]
    
df = pd.DataFrame(data)
```

```python
# Have a look at the datatypes in the table
print(df.dtypes)
```

We will save the table in a Parquet file and have a look at the data inside. This can be our first opportunity to test the runtime needed for this operation in both libraries.

```python
# save to parquet and load data using pandas
start_time_pd_pq = time.time()
df.to_parquet("data_pd_pq.parquet")

df_loaded = pd.read_parquet("data_pd_pq.parquet")
end_time_pd_pq = time.time()
time_taken_pd_pq = end_time_pd_pq - start_time_pd_pq

# Print data and runtime
print(df_loaded.head())
print(f"Time taken by pd.read_parquet(): {time_taken_pd_pq} seconds")
```

```python
# save to parquet and load data using pandas

start_time_pl_pq = time.time()
df.to_parquet("data_pl_pq.parquet")

df_loaded = pl.read_parquet("data_pl_pq.parquet")
end_time_pl_pq = time.time()
time_taken_pl_pq = end_time_pl_pq - start_time_pl_pq

# Print data and runtime
print(df_loaded.head())
print(f"Time taken by pl.read_parquet(): {time_taken_pl_pq} seconds")
```

Looks like Polars is faster and displays the table in a table bonus points.

### <font style="color:red">Reading Parquet</font> 

Here we compare the time and cpu used for reading a parquet file. In order to avoid outliers or exceptions we will run the read-parquet function 15 times and collect the runtimes for each library.

```python
runtimes_pl_pq = []

# Repeat the code 15 times
for _ in range(15):   
    
    start_time_load = time.time()
    df_loaded = pl.read_parquet("data_pl_pq.parquet")
    end_time_load = time.time()

    runtime = end_time_load - start_time_load
    runtimes_pl_pq.append(runtime)

# Print runtimes
print("Runtimes polars:", runtimes_pl_pq)
```

```python
runtimes_pd_pq = []

# Repeat the code 15 times
for _ in range(15):
    
    start_time_load = time.time()
    df_loaded = pd.read_parquet("data_pd_pq.parquet")
    end_time_load = time.time()

    runtime = end_time_load - start_time_load
    runtimes_pd_pq.append(runtime)

# Print runtimes
print("Runtimes pandas:", runtimes_pd_pq)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
import statistics as st

# Perform a t-test on the two lists
t_statistic, p_value = ttest_ind(runtimes_pd_pq, runtimes_pl_pq)

# Calculate the standard deviation of the runtimes
std_pd_pq = np.std(runtimes_pd_pq)
std_pl_pq = np.std(runtimes_pl_pq)

# Create a df
data = {
    'Lists': ['pandas runtimes', 'polars runtimes'],
    'Runtime': [st.mean(runtimes_pd_pq), st.mean(runtimes_pl_pq)],
    'Standard Deviation': [std_pd_pq, std_pl_pq]
}
df = pd.DataFrame(data)

# Plot the data
ax = sns.barplot(x='Lists', y='Runtime', data=df, yerr=df['Standard Deviation'], capsize=10)
plt.xlabel('Lists')
plt.ylabel('Runtime (sec)')
plt.title('Comparison of Parquet File Generate and Read Times between Pandas and Polars')

# Add text indicating the statistic
if p_value < 0.05:
    significance = 'Significant Difference'
else:
    significance = 'No Significant Difference'

plt.text(0.5, df['Runtime'].max() * 0.9, f'p-value: {p_value:.4f}\n{significance}',
         horizontalalignment='center', verticalalignment='top', fontweight='bold')

# Add mean value annotation
for p in ax.patches:
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()
    ax.annotate(f'{y:.3f}', (x, y), ha='center', va='bottom')


plt.savefig('plot_read_df.png')
```

### <font style="color:red">Aggreagate functions</font> 

Here we will perform some aggregate functions on the different dataframes. dfc17r1000 = A, dfc17df1000000 = B, dfc27r1000000 = C. again here we will run each aggreagate function 15 times and save the average to our runtime list for each dictionary.

```python
#the aggreagate functions we will use are
aggregate_functions = ['min', 'max', 'mean', 'median', 'std']
```

Pandas aggregate fucntions

```python
# adjust so data is saved to a file

# Select columns to include in the aggregate functions
included_columns = dfc17r1000.columns[7:]

# Initialize lists to store the runtimes and CPU usages
runtimes_agr_pd_A = []
cpu_usages_agr_pd_A = []

# Perform the aggregate functions and record the runtime and CPU usage
for function in aggregate_functions:
    function_runtimes = []
    function_cpu_usages = []
    
    for _ in range(15):
        start_time = time.time()
        cpu_usage_start = psutil.cpu_percent()
    
        result = dfc17r1000[included_columns].agg(function)
    
        end_time = time.time()
        cpu_usage_end = psutil.cpu_percent()
    
        runtime = end_time - start_time
        cpu_usage = cpu_usage_end - cpu_usage_start
    
        function_runtimes.append(runtime)
        function_cpu_usages.append(cpu_usage)
    
    average_runtime = sum(function_runtimes) / len(function_runtimes)
    average_cpu_usage = sum(function_cpu_usages) / len(function_cpu_usages)
    
    runtimes_agr_pd_A.append(average_runtime)
    cpu_usages_agr_pd_A.append(average_cpu_usage)
    
    print(f"Aggregate function: {function}")
    print(result)
    print(f"Average Runtime: {average_runtime:.6f} seconds")
    print(f"Average CPU Usage: {average_cpu_usage:.3f}%\n")
    
runtimes_agr_pd_A = [round(usage, 6) for usage in runtimes_agr_pd_A]
cpu_usages_agr_pd_A = [round(usage, 3) for usage in cpu_usages_agr_pd_A]


# Save the runtimes and CPU usages to variables
min_runtime, max_runtime, mean_runtime, median_runtime, std_runtime = runtimes_agr_pd_A
min_cpu_usage, max_cpu_usage, mean_cpu_usage, median_cpu_usage, std_cpu_usage = cpu_usages_agr_pd_A
```

Now we save the resultst to a csv

```python
# Pandas df for data per tablesize
pd_agr_dfc17r1000 = pd.DataFrame({
    'Runtimes':runtimes_agr_pd_A, 
    'CPUsage':cpu_usages_agr_pd_A,
    'functions':aggregate_functions
})
pd_agr_dfc17r1000000 = pd.DataFrame({
    'Runtimes':runtimes_agr_pd_B, 
    'CPUsage':cpu_usages_agr_pd_B,
    'functions':aggregate_functions
})
pd_agr_dfc27r1000000 = pd.DataFrame({
    'Runtimes':runtimes_agr_pd_C, 
    'CPUsage':cpu_usages_agr_pd_C,
    'functions':aggregate_functions,
})

# save to .csv
pd_agr_dfc17r1000.to_csv('pd_agr_dfc17r1000.csv', index=False)
pd_agr_dfc17r1000000.to_csv('pd_agr_dfc17r1000000.csv', index=False)
pd_agr_dfc27r1000000.to_csv('pd_agr_dfc27r1000000.csv', index=False)
```
